```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = 'https://movieweb.com/movies/2020/'

user_agent = {'User-agent': 'Mozilla/5.0'}

response  = requests.get(url, headers = user_agent)

page = response.text

soup = BeautifulSoup(page,"lxml")
```

```python
movie_items = soup.find(class_='movie-items').find_all('a',class_='movie-wrap',href=True)
movie_items[0]
```

```python
movie_title = []

for title in soup.find(class_='movie-items').find_all('p',class_='movie-title'):
    movie_title.append(title.text)
    
len(movie_title)
```

```python
links = []

for link in soup.find(class_='movie-items').find_all('a',class_='movie-wrap',href=True):
    links.append(link['href'])

links
```

```python
movie_links = list(zip(movie_title,links))
```
