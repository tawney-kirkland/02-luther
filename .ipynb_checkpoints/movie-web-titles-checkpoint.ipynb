```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = 'https://movieweb.com/movies/2020/'

user_agent = {'User-agent': 'Mozilla/5.0'}

response  = requests.get(url, headers = user_agent)

page = response.text

soup = BeautifulSoup(page,"lxml")
```

```python
# Drop unnecessary elements
soup.find('div',class_='poster').decompose()
```

```python
# Filter out movies that do not have links
movies_with_links = soup.find(class_='movie-items').find_all('a',class_='movie-wrap',href=True)
```

```python
# Find movie titles
movie_titles = [element.text for element in movies_with_links]
movie_titles = [element.split('\n')[2].strip() for element in movie_titles]
```

```python
#Find movie links
links = []

for link in soup.find(class_='movie-items').find_all('a',class_='movie-wrap',href=True):
    links.append(link['href'])

links
```

```python
#Create compiled list of movie titles and associated links
movie_links = list(zip(movie_titles,links))
```
